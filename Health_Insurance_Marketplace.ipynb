```python
import pandas as pd
from sqlalchemy import create_engine 
import psycopg2
```

```python
# Read CSV files into DataFrames
df1a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Benefits_Cost_Sharing_PUF.csv')
df1b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Benefits_Cost_Sharing_PUF.csv')
df1c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\Benefits_Cost_Sharing_PUF_2015-12-08.csv', encoding='latin-1')

# Concatenate the DataFrames
df1 = pd.concat([df1a, df1b, df1c], ignore_index=True)

def process_dataframe(df1, columns_to_drop):
    # Calculate the threshold for dropping columns based on non-null values
    threshold = len(df1) * 0.8
    
    # Drop columns with more than 20% missing values
    processed_df = df1.dropna(thresh=threshold, axis=1)
    
    # Drop duplicate rows
    processed_df = processed_df.drop_duplicates()
    
    # Drop specified columns
    processed_df = processed_df.drop(columns=columns_to_drop)
    
    # Convert 'ImportDate' column to datetime format
    processed_df['ImportDate'] = pd.to_datetime(processed_df['ImportDate'].apply(lambda x: x.split(' ')[0]), format='mixed')
    
    # Return the processed DataFrame
    return processed_df

# List of column names to drop
columns_to_drop = ['IssuerId2', 'StateCode2','IsExclFromInnMOOP', 'IsExclFromOonMOOP']

# Call the function to process the DataFrame
Benefits_Cost = process_dataframe(df1, columns_to_drop)
```

```python
# Read CSV files into DataFrames
df2a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Business_Rules_PUF.csv')
df2b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Business_Rules_PUF_Reformat.csv')
df2c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\Business_Rules_PUF_2015-12-08.csv', encoding='latin-1')

# Concatenate the DataFrames
df2 = pd.concat([df2a, df2b, df2c], ignore_index=True)

# List of column names to drop
columns_to_drop = ['IssuerId2', 'EnrolleeContractRateDeterminationRule']

# Call the function to process the DataFrame
Business_Rules = process_dataframe(df2, columns_to_drop)
```

```python
# Read CSV files into DataFrames
df3a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Network_PUF.csv')
df3b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Network_PUF.csv')
df3c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\Network_PUF_2015-12-08.csv', encoding='latin-1')
# Concatenate the DataFrames
df3 = pd.concat([df3a, df3b, df3c], ignore_index=True)

# List of column names to drop
columns_to_drop = ['IssuerId2', 'StateCode2','NetworkURL']

# Call the function to process the DataFrame
Network = process_dataframe(df3, columns_to_drop)
```

```python
# Read CSV files into DataFrames
df4a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Plan_Attributes_PUF_2014_2015-03-09.csv')
df4b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Plan_Attributes_PUF.csv')
df4c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\Plan_Attributes_PUF_2015-12-08.csv', encoding='latin-1')

# Concatenate the DataFrames
df4 = pd.concat([df4a, df4b, df4c], ignore_index=True)

# List of column names to drop
columns_to_drop = ['IssuerId2', 'StateCode2', 'InpatientCopaymentMaximumDays', 'QHPNonQHPTypeId',
                   'MultipleInNetworkTiers', 'TEHBCombInnOonIndividualMOOP', 'TEHBOutOfNetIndividualMOOP',
                   'FormularyURL', 'URLForSummaryofBenefitsCoverage','PlanBrochure',
                   'BeginPrimaryCareDeductibleCoinsuranceAfterNumberOfCopays',
                   'BeginPrimaryCareCostSharingAfterNumberOfVisits']

# Call the function to process the DataFrame
Plan_Attributes = process_dataframe(df4, columns_to_drop)
```

```python
df5a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Rate_PUF.csv')
df5b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Rate_PUF.csv')
df5c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\Rate_PUF_2015-12-08.csv',encoding='latin-1')

df5 = pd.concat([df5a, df5b, df5c], ignore_index=True)
# List of column names to drop
columns_to_drop = ['IssuerId2']

# Call the function to process the DataFrame
Rate = process_dataframe(df5, columns_to_drop)
```

```python
df6a = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2014\\Service_Area_PUF.csv')
df6b = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2015\\Service_Area_PUF.csv')
df6c = pd.read_csv('C:\\Users\\ronel\\Desktop\\health insurance\\2016\\ServiceArea_PUF_2015-12-08.csv', encoding='latin-1')

df6 = pd.concat([df6a, df6b, df6c], ignore_index=True)

# List of column names to drop
columns_to_drop = ['IssuerId2', 'StateCode2']

# Call the function to process the DataFrame
Service_Area = process_dataframe(df6, columns_to_drop)
```

```python
database_url = 'postgresql://postgres:postgres@localhost:5432/Health_Insurance_Marketplace'

# Create a SQLAlchemy engine
engine = create_engine(database_url)


table_name1 = 'Benefits_Cost'
table_name2 = 'Business_Rules'
table_name3 = 'Network'
table_name4 = 'Plan_Attributes'
table_name5 = 'Rate'
table_name6 = 'Service_Area'


Benefits_Cost.to_sql(table_name1, con=engine, index=False, if_exists='append')
Business_Rules.to_sql(table_name2, con=engine, index=False, if_exists='append')
Network.to_sql(table_name3, con=engine, index=False, if_exists='append')
Plan_Attributes.to_sql(table_name4, con=engine, index=False, if_exists='append')
Rate.to_sql(table_name5, con=engine, index=False, if_exists='append')
Service_Area.to_sql(table_name6, con=engine, index=False, if_exists='append')
```
